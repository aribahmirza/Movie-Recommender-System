```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj): # Fixed indentation here
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L # Fixed indentation here

```

```python
 #only fetch direector from crew
 def fetch_direct(obj):
   L = []
   for i in ast.literal_eval(obj):
       if i['job']=='Director':
           L.append(i['name'])
           break
   return L
```

```python
movie['crew']=movie['crew'].apply(fetch_direct)
```

```python
import ast

# Handle potential float values in 'overview' column
def split_overview(x):
    if isinstance(x, str):
        return x.split()
    else:
        return []

movie['overview'] = movie['overview'].apply(split_overview)
```

```python
#remove space
#James Cameron=JamesCameron
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['	keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movie.head()
```

```python
#make new column name as tag and combine all columns
movie['tags']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
movie.head()
```

```python
from nltk.stem.porter  import PorterStemmer
ps=PorterStemmer()
```

```python
 #helper function
 def stem(text):
   y=[]
   for i in text.split():
    y.append(ps.stem(i))

   return " ".join(y)
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector)


```

```python
similarity=cosine_similarity(vector)
similarity.shape
```

```python
def  recommend(movie):
  movie_index=newdf[newdf['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:

    print(newdf.iloc[i[0]].title)
recommend('Planet 51')
```
